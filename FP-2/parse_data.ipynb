# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('flow.csv')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- du/dx
- du/dy
- dv/dy
- du/dt
- dv/dt

**IVs**:
- t time
- x position
- y position
- u x speed (dx/dt)
- v y speed (dy/dt)
- p pressure
- sample_id labeling a dataset
- flow_type “laminar” / “turbulent” (this CSV is not displayed somehow)



<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['t', 'u', 'v', 'dudt', 'dvdt']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'t':'time'} )

df.describe()
```
