### Figure 1

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

plt.figure(figsize=(6, 4))
plt.plot(a, 'ko-', linewidth=1.5, markersize=6)
plt.xlabel('Time') 
plt.ylabel('Value') 
plt.grid(False)
plt.tight_layout()
plt.show()

```

### Figure 2

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig2.npy' )

plt.figure(figsize=(6, 4))
plt.plot(a[:, 0], color='red',   label='Red')
plt.plot(a[:, 1], color='green', label='Green')
plt.plot(a[:, 2], color='blue',  label='Blue')

plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()

plt.tight_layout()
plt.show()
```

### Figure 3

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig3.npy' )

plt.figure(figsize=(6, 4))
plt.scatter(
    a[:, 0], 
    a[:, 1], 
    s=200,  
    color='blue',  
    alpha=0.4,     
    edgecolors='none' 
)

plt.xlabel('X value')
plt.ylabel('Y value')

plt.tight_layout()
plt.show()
```

### Figure 4

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig4.npy' )

labels = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(6, 4))
plt.bar(labels, a, color='gray')  # 

plt.xlabel('Category')
plt.ylabel('Value')

plt.ylim(0.0, 0.5)

plt.tight_layout()
plt.show()
```

### Figure 5

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig5.npy' )

plt.figure(figsize=(6, 4))

img = plt.imshow(
    a, 
    cmap='RdBu',     
    origin='lower',     
    aspect='auto',        
    vmin=-1.0, vmax=1.0  
)

plt.xticks(np.arange(0, 41, 5))   
plt.yticks(np.arange(0, 121, 20)) 

cbar = plt.colorbar(img)

plt.tight_layout()
plt.show()
```
