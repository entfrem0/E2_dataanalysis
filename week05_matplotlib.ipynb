## List comprehension

A list comprehension is a Python programming technique that allows you tu use a for loop to create a list in a single-line command

```python
x = [2*i for i in [0,1,2,3]]
print(x)
```

A list comprehension is short-hand (short-form coding) for a full for loop:

```python
x = []
for i in [0,1,2,3]:
    x.append(2*i)
    
print(x)
```

List comprehensions are useful for plotting data

```python
import numpy as np

np.random.seed(0) # controls random number generation
x0 = np.random.randn(10) # 10 random values from the normal distribution
x1 = np.random.randn(10) # 10 random values from the normal distribution
x2 = np.random.randn(10) # 10 random values from the normal distribution

print(x0)


plt.figure()
ax = plt.axes()
ax.plot(x0)
ax.plot(x1)
ax.plot(x2)
plt.show()

```

```python
plt.figure()
ax = plt.axes()
[ax.plot(xx) for xx in [x0,x1,x2]]
plt.show()
```
