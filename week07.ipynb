## For loops

For loops allow you to do similar, repetitive calculations on several or many instances of data.

for loops concepts:

- basic for loop
- list comprehensions
- zip
- enumerate
- enumerate-zip

### Basic for loop

````python
for i in [0,5,6,8]:
    y = i + 10
print(y)
````

````python
for i in range(5):
    print(i)
    
````

````python
x = list(range(5))
print(x)

x = [0,1,2,3,4]
print(x)
````

### List comprehensions

Doing the same hting with a full for loop requires substantially more code

````python
x = []
for i in [0,5,6,8]:
    y = i + 10
    x.append(y)
    
print(x)
````

### zip

The `zip` function allows you to comine two or more iterable objects in a for loop

````python
x0 = [0,5,8]
x1 = [10,20,30]

for y0,y1 in zip(x0,x1):
    print(y0,y1)
````

### enumerate

The `enumerate` function allows you to get the index for each element of an iterable for each iteration of the for loop

````python
x = [0,5,8,20]
for i,xx in enumerate(x):
    print(1,xx)
````

### enumerate-zip

You can combine `enumerate` and `zip` to create for loops:

````python
x0 = [0,5,8]
x1 = [10,20,30]

for i, (y0,y1) in enumerate(zip(x0,x1)):
    print(1,y0,y1)
````

# Scripts

A script is an ASCII files (i.e., a plain text file) that has a .py extension.

Scripts are meant to achieve a specific computational goal.

For example: calculate summary statistics for a specific dataset.

In Jupyter, scripts can be executed using the magic run command `%run`

`%run myscript.py`

# Modules

Like scripts, modules are ASCII files that have .py extentions.

Module are meant to contain general code that can be re-used in several scripts OR as a convenient way to hide computation details.

Modules need to be imported to be used.

Example:
```python
import mymodule as mm
mm.run_descriptive_stats()
```
