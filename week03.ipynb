## Vectors in Numpy

NumPy does NOT distinguish between row and column vectors like in linear (matrix) argebra.

By conevntion, all vectors should be 1D arrays

```python
import numpy as np

x = np.array([ [5], [2], [1]])
print(x)
print()

x= np.array([ [5,2,1]])

print(x)
print()

x = x.T #transpose
print(x)
```
